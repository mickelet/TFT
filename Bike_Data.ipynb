```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

top_10000 = df.sort_values(by='Cost', ascending=False).head(1000000)

top_10000['Date'] = pd.to_datetime(df['Date'])


monthly_trend = df.groupby(['Date', 'Product_Description'])['Cost'].sum().reset_index()

sub_categories = sorted(monthly_trend['Product_Description'].unique())
label_mapping = {sub_cat: str(i+1) for i, sub_cat in enumerate(sub_categories)}
monthly_trend['Sub_Category_Label'] = monthly_trend['Product_Description'].map(label_mapping)

plt.figure(figsize=(10, 6))
for label, group in monthly_trend.groupby('Sub_Category_Label'):
    plt.plot(group['Date'], group['Cost'], label=label)
    # Add label at the end of each line
    plt.text(group['Date'].iloc[-1], group['Cost'].iloc[-1], label)


plt.title('Monthly Trend of Profit by Sub-Category')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.legend(title='Sub-Category', loc = 'upper left')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

top_10000 = df.sort_values(by='Cost', ascending=False).head(1000000)

top_10000['Date'] = pd.to_datetime(df['Date'])


monthly_trend = df.groupby(['Date', 'Product_Description'])['Cost'].sum().reset_index()

sub_categories = sorted(monthly_trend['Product_Description'].unique())
label_mapping = {sub_cat: f"{i+1}: {sub_cat}" for i, sub_cat in enumerate(sub_categories)}
monthly_trend['Sub_Category_Label'] = monthly_trend['Product_Description'].map(label_mapping)

plt.figure(figsize=(10, 6))
for label, group in monthly_trend.groupby('Sub_Category_Label'):
    plt.plot(group['Date'], group['Cost'], label=label)
    # Add label at the end of each line
    plt.text(group['Date'].iloc[-1], group['Cost'].iloc[-1].round(2), label.split(":")[0])


plt.title('Monthly Trend of Profit by Product Description')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.legend(title='Product Description', loc = 'upper left')
plt.tight_layout()
plt.show()
```

Question 3: Customer Demographics Analysis


What is the distribution of customers by age group and gender?

Which age groups contribute the most to total revenue?

How do purchasing patterns differ between male and female customers?

```python
display(df.head())
```

```python
gender_age_counts = df.groupby(['Customer_Gender', 'Age_Group']).size().reset_index(name='Count')
pivot_table = gender_age_counts.pivot(index='Age_Group', columns='Customer_Gender', values='Count').fillna(0).astype(int)

print(pivot_table)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

pivot_table.plot(kind='bar', stacked=True)
plt.title('Distribution of Customers by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Pie chart for Age Group distribution
age_group_counts = df['Age_Group'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(age_group_counts, labels=age_group_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Customers by Age Group')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
